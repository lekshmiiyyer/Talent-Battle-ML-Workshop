```python
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
x=np.array([100,150,200,250,300]).reshape(-1,1) #house sizes in sqft
y=np.array([250000,350000,450000,550000,650000]) #house prices
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
new_house_size=np.array([175,225]).reshape(-1,1)
predicted_prices=model.predict(new_house_size) #predict the prizes for same bassed on trained model
```

```python
print("Intercept=",model.intercept_)
print("slope=",model.coef_[0]) #y=b0+b1x+Error
```

```python
plt.scatter(x,y,color='blue',label='Data')
plt.plot(x,model.predict(x),color='red',label='Regression Line')
plt.scatter(new_house_size,predicted_prices,color='green',label='Predicted Prices')
plt.xlabel('House Size')
plt.ylabel('House Price')
plt.legend()
plt.title('sample example for linear regression')
plt.show()
```
